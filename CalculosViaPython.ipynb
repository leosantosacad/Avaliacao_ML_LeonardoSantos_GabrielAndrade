# **Início do código**


O código usa um laço de repetição (for) para analisar cada produto separadamente. Para cada tipo de tinta, ele faz duas coisas principais:

​Calcula a Confiabilidade (R²): Ele avalia a precisão da nossa linha de tendência, dando uma "nota" (de 0 a 1) para sabermos se a previsão matemática para o futuro é segura. ​Desenha o Gráfico: Ele pega os números e transforma em imagem, colocando os pontos azuis (histórico de vendas reais) e traçando a linha vermelha (a tendência para os próximos meses).

```python
plt.figure(figsize=(15, 10))
produtos = df['Produto'].unique()

for idx, produto in enumerate(produtos, 1):
    df_prod = df[df['Produto'] == produto]
    X = df_prod[['Mês']]
    y = df_prod['Vendas']

    modelo = LinearRegression()
    modelo.fit(X, y)

    X_futuro = np.array(list(X['Mês']) + [13, 14, 15]).reshape(-1, 1)
    y_previsto = modelo.predict(X_futuro)

    # R² e Equação
    r2 = r2_score(y, modelo.predict(X)) if len(y) > 1 else 1.0
    a = modelo.coef_[0]
    b = modelo.intercept_
    print(f"{produto} -> y = {a:.2f}x + {b:.2f} | R² = {r2:.4f}")

    plt.subplot(3, 2, idx)
    plt.scatter(X, y, color='blue', label='Vendas Reais')
    plt.plot(X_futuro, y_previsto, color='red', label='Linha de Tendência')
    plt.title(f'{produto} (R²={r2:.2f})')
    plt.xlabel('Mês')
    plt.ylabel('Vendas')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def pegar_trimestre(mes):
    if mes <= 3: return 1
    elif mes <= 6: return 2
    elif mes <= 9: return 3
    elif mes <= 12: return 4
    else: return 5 # Meses 13, 14 e 15

resultados_trimestre = []
produtos = df['Produto'].unique()

for produto in produtos:
    df_prod = df[df['Produto'] == produto]
    X = df_prod[['Mês']]
    y = df_prod['Vendas']

    modelo = LinearRegression()
    modelo.fit(X, y)

    for mes in range(1, 16):
        trimestre = pegar_trimestre(mes)

        venda_real = df_prod[df_prod['Mês'] == mes]['Vendas'].values
        venda_real = venda_real[0] if len(venda_real) > 0 else 0

        venda_prevista = modelo.predict([[mes]])[0]

        resultados_trimestre.append({
            'Produto': produto,
            'Trimestre': trimestre,
            'Vendas Reais': venda_real,
            'Vendas Previstas': venda_prevista
        })

df_trimestres = pd.DataFrame(resultados_trimestre)
resumo_final = df_trimestres.groupby(['Produto', 'Trimestre']).sum().reset_index()

print("--- RESULTADOS POR TRIMESTRE ---\n")
for produto in produtos:
    print(f"**{produto}**")
    df_resumo_prod = resumo_final[resumo_final['Produto'] == produto]

    for _, linha in df_resumo_prod.iterrows():
        tri = int(linha['Trimestre'])
        real = linha['Vendas Reais']
        prev = linha['Vendas Previstas']

        if tri == 5:
            print(f"* Tri 5 (Meses 13-15): Futuro | Previsto = {prev:.2f}")
        elif real > 0:
            print(f"* Tri {tri}: Real = {real:.0f} | Previsto = {prev:.2f}")
        else:
            print(f"* Tri {tri}: Sem dados reais registrados.")
```
